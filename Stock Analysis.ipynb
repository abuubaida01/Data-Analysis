```python
import pandas as pd

import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime as dt
import seaborn as sns
%matplotlib inline
```

```python
#!pip install pandas_datareader
import pandas_datareader.data as web 
#or we can use other method as well for extracting data from Yahoo finance or google finance.
import pandas_datareader as pdr
```

```python
start = dt(2004, 1, 1)
end = dt(2022, 1, 15)
```

# Resampling Quarterly

```python
#lis = [nvda, qcom, ibm, intc]
m_nvda_18 = nvda_18.resample('4M').mean()
m_qcom_18 = qcom_18.resample('4M').mean()
m_ibm_18 = ibm_18.resample('4M').mean()
m_intc_18 = intc_18.resample('4M').mean()
```

```python
fg, ax = plt.subplots(2,2, figsize= (15, 7), sharex=True, )
ax[0,0].scatter(m_intc_18.index, m_intc_18.Close, c='y', edgecolor='k')
ax[0,0].tick_params(labelrotation=90)
ax[0,0].set_title('Intel Closing Price')

ax[0,1].scatter(m_nvda_18.index, m_nvda_18.Close)
ax[0,1].tick_params(labelrotation=90)
ax[0,1].set_title('Nvidia Closing Price')


ax[1,0].scatter(m_qcom_18.index, m_qcom_18.Close, c='g')
ax[1,0].tick_params(labelrotation=90)
ax[1,0].set_title('QUALCOMM Closing Price')

ax[1,1].scatter(m_ibm_18.index, m_ibm_18.Close, c='r')
ax[1,1].tick_params(labelrotation=90)
ax[1,1].set_title('IBM Closing Price')
```

 for getting the actual difference btw them. We need to minus close price with open price. If company has positive value so means the compnay is flourishing. 

```python
#lis = [nvda, qcom, ibm, intc]
nvda.head()
```

```python
nvda['profit'] = nvda.Open - nvda.Close
qcom['profit'] = qcom.Open - qcom.Close
ibm['profit'] = ibm.Open - ibm.Close
intc['profit'] = intc.Open - intc.Close
```

```python
w_nvda = nvda.resample('W').mean()
w_qcom = qcom.resample('W').mean()
w_ibm = ibm.resample('W').mean()
w_intc = intc.resample('W').mean()
```

```python
#plt.style.use('fivethirtyeigh')
fg, ax=plt.subplots(2,2, figsize=(15,7), sharex=True)
ax[0,0].scatter(w_intc.loc['2018, 1, 15': '2022, 1, 15'].index, w_intc.profit.loc['2018, 1, 15': '2022, 1, 15'], c='y', edgecolor='k')
ax[0,0].tick_params(labelrotation=90)
ax[0,0].set_title('Intel Progress')

ax[0,1].scatter(w_nvda.loc['2018, 1, 15': '2022, 1, 15'].index, w_nvda.profit.loc['2018, 1, 15': '2022, 1, 15'])
ax[0,1].tick_params(labelrotation=90)
ax[0,1].set_title('Nvidia progress ')


ax[1,0].scatter(w_qcom.loc['2018, 1, 15': '2022, 1, 15'].index, w_qcom.profit.loc['2018, 1, 15': '2022, 1, 15'], c='g')
ax[1,0].tick_params(labelrotation=90)
ax[1,0].set_title('QUALCOMM Progress')

ax[1,1].scatter(w_ibm.loc['2018, 1, 15': '2022, 1, 15'].index, w_ibm.profit.loc['2018, 1, 15': '2022, 1, 15'], c='r')
ax[1,1].tick_params(labelrotation=90)
ax[1,1].set_title('IBM Progress')
```
